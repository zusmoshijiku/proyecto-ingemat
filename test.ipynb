```python
from PIL import Image, ImageColor, ImageOps
import numpy as np
from matplotlib import pyplot as plt
```

```python
imagen = Image.open("0.jpg")
imagen = imagen.convert("L", colors=16)
imagen = imagen.resize((8,8))
imagen = ImageOps.posterize(imagen, bits=8)
a = np.asarray(imagen)
print(a)
```

```python
plt.imshow(a, cmap= "gray")
```

```python
aguante = Image.open("miniatura.png")
aguante = aguante.convert("L")
aguante = ImageOps.posterize(aguante, bits=4)
aguante = aguante.resize((28,28))
ag = np.asarray(aguante)
plt.imshow(ag, cmap="gray")
print(aguante)
```
